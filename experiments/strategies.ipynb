```python
df = pd.DataFrame().ta.ticker("SPY", period="6mo", interval="1d")
```

```python
df.ta.strategy("Momentum")
```

```python
df["MFI_14"] = df.ta.mfi(append=True)

df["MACD_Above"] = df["MACDh_12_26_9"] > 0
df["MACD_Below"] = df["MACDh_12_26_9"] < 0
df["MFI_Overbought"] = df["MFI_14"] > 70
df["MFI_Oversold"] = df["MFI_14"] < 30

df["MACD_CrossOver"] = (df["MACD_Above"].shift(1) - df["MACD_Above"]) == 1
df["MACD_CrossOver"] = df["MACD_CrossOver"].astype(int)
df["MACD_CrossUnder"] = (df["MACD_Above"] - df["MACD_Above"].shift(1)) == 1
df["MACD_CrossUnder"] = df["MACD_CrossUnder"].astype(int)
```

```python
print(sum(df["MFI_Oversold"]))
sum(df["MFI_Overbought"])
```

```python
import vectorbt as vbt

# signal to buy (changes from false --> true when buy, true --> false when sell)
df["SIGNAL"] = df["MFI_Oversold"] & df["MACD_CrossOver"] | df["MFI_Overbought"] & df["MACD_CrossUnder"]

print(sum(df["SIGNAL"]))

golden = df.ta.tsignals(df["SIGNAL"], asbool=True, append=True)

# Sanity Check (Ensure data exists)
print(df)

# Create the Signals Portfolio
pf = vbt.Portfolio.from_signals(df["Close"], entries=golden.TS_Entries, exits=golden.TS_Exits, freq="D", init_cash=100_000, fees=0.0025, slippage=0.0025)

# Print Portfolio Stats and Return Stats
print(pf.stats())
print(pf.returns_stats())
```

```python
import matplotlib.pyplot as plt

r = pf.daily_returns()
cash = 100000
all_vals = []
for i in range(len(r)):
    cash = cash * (1 + r[i])
    all_vals.append(cash)

print("total return", cash/100000)
print("min pf value", min(all_vals)/100000)
print("max pf value", max(all_vals)/100000)


# Plot the Portfolio Value
pf.value().vbt.plot()
# plt.show()
```

```python
# calculate duration of each trade
entries = golden.TS_Entries.astype(int)
exits = golden.TS_Exits.astype(int)

for i in range(len(entries)):
    if entries[i] == 1:
        print("entry", i)
        for j in range(i, len(exits)):
            if exits[j] == 1:
                print("exit", j)
                print("duration", j - i)
                break
```
